# Challenge 2

## Text Cleaning

```python
import re

def clean_up(s):
    pattern = '[^a-z]+'
    x = re.sub(pattern, ' ', s.lower())
    return x.split('http')[0]

s = '''@Ironhack's-#Q website 776-is http://ironhack.com [(2018)]")'''
print (clean_up(s))

#output = ironhack s  q website  is
print(s)
```

## Tokenización

```python
def tokenize(s):
    return word_tokenize(s)

tokenize(clean_up(s))
```

## Stemming and Lemmatization

```python
nltk.download('wordnet')
from nltk import stem
from nltk.stem import WordNetLemmatizer
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()


from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize
 
ps = PorterStemmer()
```

```python
def stem_and_lemmatize(l):
    lemmatizer = WordNetLemmatizer()
    stemmer = stem.PorterStemmer()
    stem_phrase = [stemmer.stem(x) for x in l] 
    stem_and_lemmatizer_phrase = [lemmatizer.lemmatize(x) for x in stem_phrase]
    return stem_and_lemmatizer_phrase

stem_and_lemmatize(tokenize(clean_up(s)))
```

## Stop Words Removal

```python
from stop_words import get_stop_words
from stop_words import safe_get_stop_words

def remove_stopwords(l):
    stop_words = get_stop_words('en')
    for i in l:
        if i in stop_words:
            l.remove(i)
    return l

remove_stopwords(stem_and_lemmatize(tokenize(clean_up(s))))
```
