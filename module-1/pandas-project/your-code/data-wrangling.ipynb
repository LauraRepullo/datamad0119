```python
import pandas as pd
import numpy as np
data = pd.read_csv('GSAF5.csv', encoding='ISO-8859-1')
```

```python
data.head(20)
```

```python
data.columns
```

```python
# saco los valores nulos por columnas, y que me muestre las columnas que tienen datos nulos:
null_cols = data.isnull().sum()
null_cols[null_cols > 0]
```

```python
#elimino las columnas que tienen mas de 5000 valores nulos:
drop_cols = list(null_cols[null_cols > 5000].index)
drop_cols
```

```python
data.columns
```

```python
#muestro el DataFrame sin las columnas que tenian mas de 5000 valores nulos:
data_without_nulls = data.drop(drop_cols, axis=1)
data_without_nulls.head()
```

```python
data_without_nulls.columns
```

```python
#elimino las columnas que no me interesan:
col_lst = ['pdf', 'href formula', 'href', 'Case Number.1', 'Case Number.2', 'Name']
data_clean1 = data_without_nulls.drop(col_lst, axis=1, inplace=False)

data_clean1.columns
```

```python
#reorganizamos las columnas para que se puedan interpretar mejor:
column_order = ['Year', 'Date', 'Time','Country', 'Area', 'Location', 'Activity','Type', 'Sex ', 'Age', 'Injury', 'Fatal (Y/N)', 'Species ','Case Number', 'original order']

data_clean1 = data_clean1[column_order]
data_clean1.head(10)
```

```python
#analizo los datos de la columna "year" para ver que datos muestra:
print(set(data_clean1['Year']))
```

```python
#elimino los registros anteriores al año 1600, porque no considero que la informacion sea veraz:
data_clean2 = data_clean1.drop(data_clean1[data_clean1['Year']< 1600].index, axis=0, inplace = False)
data_clean2
```

```python
print(set(data_clean2['Fatal (Y/N)']))
```

```python
#homogeneizamos los datos de la columna Fatal (Y/N):
data_clean2['Fatal (Y/N)'] = data_clean2['Fatal (Y/N)'].str.replace('n', 'N')
data_clean2['Fatal (Y/N)'] = data_clean2['Fatal (Y/N)'].str.replace('F', 'UNKNOWN')
data_clean2['Fatal (Y/N)'] = data_clean2['Fatal (Y/N)'].str.replace('N ', 'N')
data_clean2['Fatal (Y/N)'] = data_clean2['Fatal (Y/N)'].str.replace(' N', 'N')
data_clean2['Fatal (Y/N)'] = data_clean2['Fatal (Y/N)'].str.replace('#VALUE!', 'UNKNOWN')
data_clean2['Fatal (Y/N)'].fillna(value='UNKNOWN', inplace=True)
print(set(data_clean2['Fatal (Y/N)']))
```

```python
#incluyo una columna con el numero de accidentes:

data_clean2['N_Attacks'] = data_clean2['original order'].count()/len(data_clean2['original order'])
data_clean2.head(10)
```

```python
#Ordenar por años/pais/area en orden descendente:
final_dataclean = data_clean2.sort_values(by=['Year', 'Country', 'Area'], ascending=False)
final_dataclean
```

```python
#calculamos los ataques registrados por paises, ordenado  de mayor a menor:
Attacks_per_Country = data_clean2.groupby(['Country'])['N_Attacks'].sum()
Attacks_per_Country
Attacks_per_Country_ordered = Attacks_per_Country.sort_values(ascending=False)
Attacks_per_Country_ordered.head(10)
```

```python
#Analizamos el numero de ataques por pais y por año:
SharkAttacks_per_Country_Year = data_clean2.groupby(['Country','Year'])['N_Attacks'].sum()
SharkAttacks_per_Country_Year_ordered = SharkAttacks_per_Country_Year.sort_values(ascending=False)
SharkAttacks_per_Country_Year_ordered
```

```python
#Analizamos los casos de fallecimiento en los ataques(en porcentaje):
SharkAttack_result_percentage= (data_clean2.groupby(['Fatal (Y/N)'])['N_Attacks'].sum())/(data_clean2['N_Attacks'].sum())*100
SharkAttack_result_percentage
```
