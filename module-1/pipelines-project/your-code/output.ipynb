```python
# Suicide Rates Overview 1985 to 2016 
# About this file: country, year, sex, age group, count of suicides, population, suicide rate, country-year composite key, HDI for year, gdp_for_year, gdp_per_capita, generation (based on age grouping average).
# https://www.kaggle.com/russellyates88/suicide-rates-overview-1985-to-2016
# HDI (Human Development Index: El índice de desarrollo humano (IDH) es un indicador del desarrollo humano por país, elaborado por el Programa de las Naciones Unidas para el Desarrollo (PNUD). 
#El índice de Desarrollo Humano (IDH) es un indicador sintético de los logros medios obtenidos en las dimensiones fundamentales del desarrollo humano:
#-tener una vida larga y saludable, 
#-adquirir conocimientos y 
#-disfrutar de un nivel de vida digno. 
#El IDH es la media aritmética de los índices normalizados de cada una de las tres dimensiones.
#Gross Domestic Product - GDP 

#1) Importamos la Dataset:

import pandas as pd

data = pd.read_csv("suicide-rates-overview-1985-to-2016.csv")

data
```

```python
#2) Limpiamos la Dataset: empezamos buscando los duplicados.
data_no_dupl = data.drop_duplicates()
data_no_dupl.describe
```

```python
#vemos que la columna "HDI for year" tiene mas de 19000 registros nulos: asi que la eliminamos:
null_cols=data_no_dupl.isnull().sum()
null_cols
```

```python
data_no_dupl1=data_no_dupl.drop(['HDI for year'], axis=1)
print(data_no_dupl1.head(10))
```

```python
data_no_dupl1.columns
```

```python
#eliminamos la columna 'country-year' porque tenemos ya otras 2 columnas que nos facilitan esos datos:
data_clean=data_no_dupl1.drop(['country-year'], axis=1)
```

```python
#creo una tabla que incluya el total de suicidios y poblacion por pais; el problema es que los datos suman los 31 años:
data_selected= data_clean[['country', 'suicides_no','population']]
data_selected_grouped= data_selected.groupby(['country']).sum()
data_selected_grouped.head()
```

```python
#para resolver el problema de la poblacion, inserto una columna que calcule una media de poblacion por pais:
data_selected_grouped['population(mean)'] = data_selected_grouped['population']/31
data_selected_grouped['suicides_no(mean)'] = data_selected_grouped['suicides_no']/31
#data_selected_grouped['suicides_no/100k pop'] = data_selected_grouped['suicides_no']/['population(mean)']
data_selected_gr= data_selected_grouped[['suicides_no(mean)', 'population(mean)']]
data_selected_gr
```

```python
#incluyo una columna con la media de poblacion de cada pais, de los años de 1985 a 2016:
data_selected_grouped['suicides_no/100k pop']= data_selected_grouped['suicides_no(mean)']/data_selected_grouped['population(mean)']*100000
suicides_per_country_ord=data_selected_grouped.sort_values(by=['suicides_no/100k pop'], ascending=False)
suicides_per_country_ord['suicides_no/100k pop'].head()
```
