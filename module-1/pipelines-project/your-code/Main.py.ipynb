```python
# Suicide Rates Overview 1985 to 2016 
# About this file: country, year, sex, age group, count of suicides, population, suicide rate, country-year composite key, HDI for year, gdp_for_year, gdp_per_capita, generation (based on age grouping average).
# https://www.kaggle.com/russellyates88/suicide-rates-overview-1985-to-2016
# HDI (Human Development Index: El índice de desarrollo humano (IDH) es un indicador del desarrollo humano por país, elaborado por el Programa de las Naciones Unidas para el Desarrollo (PNUD). 
#El índice de Desarrollo Humano (IDH) es un indicador sintético de los logros medios obtenidos en las dimensiones fundamentales del desarrollo humano:
#-tener una vida larga y saludable, 
#-adquirir conocimientos y 
#-disfrutar de un nivel de vida digno. 
#El IDH es la media aritmética de los índices normalizados de cada una de las tres dimensiones.
#Gross Domestic Product - GDP 

#1) Importamos la Dataset:

import pandas as pd

data = pd.read_csv("suicide-rates-overview-1985-to-2016.csv")

data
```

```python
#2) Limpiamos la Dataset: empezamos buscando los duplicados.
data_without_duplicates = data.drop_duplicates()
data_without_duplicates.describe
```

```python
#eliminamos la columna "HDI for year", ya que tiene mas de 19000 registros nulos:
null_cols=data_without_duplicates.isnull().sum()
null_cols
```

```python
data_without_duplicates=data_without_duplicates.drop(['HDI for year'], axis=0)
print(data_without_duplicates)
```

```python
data_without_duplicates.columns
```

```python
#eliminamos la columna 'country-year' porque tenemos ya otras 2 columnas que nos facilitan esos datos:
data_without_duplicates=data_without_duplicates.drop(['country-year'], axis=1)
```

```python
data_without_duplicates.columns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.cbook as cbk
```

```python
#calculamos el numero de suicidios en funcion del sexo (en porcentaje)
Suicides_per_sex= data_without_duplicates.groupby(['sex'])['suicides/100k pop'].sum()/(75014.77+281529.06)*100
Suicides_per_sex
```

```python
#calculamos suicidios por pais:
Suicides_per_country= data_without_duplicates.groupby(['country'])['suicides/100k pop'].sum()
Suicides_per_country_ordered = Suicides_per_country.sort_values(ascending=False)
Suicides_per_country_ordered.head(5)
```

```python
Suicides_per_year= data_without_duplicates.groupby(['year','country'])['suicides/100k pop'].sum()
Suicides_per_year.head(10)
```

```python
#Pivotamos la columna año, para conseguir una tabla de la evolucion del numero de casos de suicidio registrados entre 1985 y 2016 por pais:
suicides_per_year_country_pivot=data_without_duplicates.pivot_table(index='country', columns='year', values='suicides/100k pop')
suicides_per_year_country_pivot.fillna(0)
```
